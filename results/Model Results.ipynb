```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2*np.pi*t)
```

```python
plt.plot(t, s)

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.grid(True)
plt.savefig("test.png")
plt.show()
```

```python
max_length = np.array ([5, 10, 15, 20, 25, 30, 35, 40])
length_acc = np.array ([0.757117, 0.818823, 0.834410, 0.836151, 0.837453, 0.840005, 0.839819, 0.841595])
length_f1 = np.array ([0.652066, 0.756395, 0.778605, 0.778331, 0.790043, 0.795169, 0.796153, 0.795941])
```

```python
plt.plot (max_length, length_acc, 'r-', label='accuracy')
plt.plot ( max_length, length_f1, 'b-', label='f1')
plt.xlabel ('max length')
plt.title ('Maximum Sentence Length Hyperparameter Search')
plt.savefig("max_length.png", bbox_inches="tight")
plt.legend ()
```
